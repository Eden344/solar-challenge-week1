```python
import pandas as pd
import os

DATA_PATH = "../data/togo-dapaong_qc.csv"

df = pd.read_csv(DATA_PATH)
df.head()
```

```python
df.columns
```

```python
# Basic info and summary
df.info()

# Summary stats for numeric columns
display(df.describe())

# Missing values per column
df.isna().sum()
```

```python
df.drop(columns=['Comments'], inplace=True)
```

```python
irradiance_cols = ['GHI', 'DNI', 'DHI', 'ModA', 'ModB']
wind_cols = ['WS', 'WSgust', 'WSstdev']

for col in irradiance_cols + wind_cols:
    df.loc[df[col] < 0, col] = 0
```

```python
from scipy import stats
import numpy as np

# Columns to check for outliers
cols_to_check = ['GHI', 'DNI', 'DHI', 'ModA', 'ModB', 'WS', 'WSgust']

# Compute Z-scores
z_scores = np.abs(stats.zscore(df[cols_to_check]))

# Count how many are beyond |Z| > 3
outliers = (z_scores > 3).sum()
outliers
```

```python
# Boolean mask of outliers
outlier_mask = (z_scores > 3).any(axis=1)

# Show rows that are outliers in any column
df_outliers = df[outlier_mask]
df_outliers.head()  # look at first few
```

```python
from scipy import stats
import numpy as np

cols_to_check = ['GHI', 'DNI', 'DHI', 'ModA', 'ModB', 'WS', 'WSgust']

# Compute Z-scores
z_scores = np.abs(stats.zscore(df[cols_to_check]))

# Create a mask for rows with any |Z| > 3
outlier_mask = (z_scores > 3).any(axis=1)

# Add a column to flag them
df['Outlier'] = outlier_mask
```

```python
for col in cols_to_check:
    median_val = df[col].median()
    df.loc[np.abs(stats.zscore(df[col])) > 3, col] = median_val
df_cleaned = df.copy()
```

```python
# Replace placeholders with NaN (future-proof way)
df_cleaned['BP'] = df_cleaned['BP'].replace(1002, np.nan)

# Fill missing values with median
df_cleaned['BP'] = df_cleaned['BP'].fillna(df_cleaned['BP'].median())
df_cleaned.isna().sum()
```

```python
import os

os.makedirs('data', exist_ok=True)
df_cleaned.to_csv('data/togo_clean.csv', index=False)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Use a safe built-in style
plt.style.use('ggplot')
```

```python
country = "togo"
df_cleaned = pd.read_csv(f"data/{country}_clean.csv", parse_dates=['Timestamp'])
```

```python
plt.figure(figsize=(14,6))
plt.plot(df_cleaned['Timestamp'], df_cleaned['GHI'], label='GHI', color='orange')
plt.plot(df_cleaned['Timestamp'], df_cleaned['DNI'], label='DNI', color='blue')
plt.plot(df_cleaned['Timestamp'], df_cleaned['DHI'], label='DHI', color='green')
plt.plot(df_cleaned['Timestamp'], df_cleaned['Tamb'], label='Tamb (C)', color='red')
plt.xlabel('Timestamp')
plt.ylabel('Value')
plt.title(f"{country.capitalize()} Solar Data Time Series")
plt.legend()
plt.show()
```

```python
# WS vs GHI
plt.figure(figsize=(8,5))
plt.scatter(df_cleaned['WS'], df_cleaned['GHI'], alpha=0.5)
plt.xlabel('Wind Speed (WS)')
plt.ylabel('GHI')
plt.title(f'{country.capitalize()} WS vs GHI')
plt.show()

# RH vs Tamb
plt.figure(figsize=(8,5))
plt.scatter(df_cleaned['RH'], df_cleaned['Tamb'], alpha=0.5, color='purple')
plt.xlabel('Relative Humidity (RH)')
plt.ylabel('Ambient Temp (Tamb)')
plt.title(f'{country.capitalize()} RH vs Tamb')
plt.show()
```

```python
cols_corr = ['GHI','DNI','DHI','ModA','ModB','Tamb','RH','WS','WSgust']
plt.figure(figsize=(10,8))
sns.heatmap(df_cleaned[cols_corr].corr(), annot=True, cmap='coolwarm')
plt.title(f'{country.capitalize()} Correlation Heatmap')
plt.show()
```

```python
df_cleaned[['GHI','WS']].hist(bins=30, figsize=(10,4))
plt.suptitle(f'{country.capitalize()} Distribution of GHI and WS')
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.scatter(
    df_cleaned['Tamb'],       # x-axis
    df_cleaned['GHI'],        # y-axis
    s=df_cleaned['RH'],       # bubble size = Relative Humidity
    alpha=0.5,
    c=df_cleaned['RH'],       # optional: color by RH
    cmap='viridis'
)
plt.xlabel('Ambient Temperature (Tamb C)')
plt.ylabel('Global Horizontal Irradiance (GHI)')
plt.title('Bubble Chart: GHI vs Tamb (bubble = RH)')
plt.colorbar(label='Relative Humidity (RH)')
plt.show()
```

```python
import numpy as np

# Convert wind direction to radians
theta = np.deg2rad(df_cleaned['WD'])
r = df_cleaned['WS']

plt.figure(figsize=(8,8))
ax = plt.subplot(111, polar=True)
hb = ax.scatter(theta, r, c=r, cmap='plasma', alpha=0.6)

ax.set_theta_zero_location('N')  # 0 = North
ax.set_theta_direction(-1)       # clockwise
plt.title('Wind Rose: WS & WD')
plt.colorbar(hb, label='Wind Speed (WS)')
plt.show()
```

```python
import seaborn as sns

plt.figure(figsize=(10,5))
sns.boxplot(x='Cleaning', y='ModA', data=df_cleaned)
plt.title('ModA: Pre vs Post Cleaning')
plt.xlabel('Cleaning (0=Pre, 1=Post)')
plt.ylabel('ModA Reading')
plt.show()

plt.figure(figsize=(10,5))
sns.boxplot(x='Cleaning', y='ModB', data=df_cleaned)
plt.title('ModB: Pre vs Post Cleaning')
plt.xlabel('Cleaning (0=Pre, 1=Post)')
plt.ylabel('ModB Reading')
plt.show()
```
